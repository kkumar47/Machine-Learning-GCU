**Part 2: Find the Minima and plot the Contour plot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize two numpy arrays with all the values that has been calculated in the Assignment
w1 = np.array([2,1.6,1.04])
w2 = np.array([-2,-0.4,-0.32])
```

```python
# Create a meshgrid of the two variables which will be used to plot the contour plot
W1, W2 = np.meshgrid(w1,w2)
```

```python
W1
```

```python
W2
```

```python
# Define the function
Y = (2*np.square(W1)) + (2*W1*W2) + (5*np.square(W2))
```

```python
Y
```

```python
# Create a Contour plot
cp = plt.contour(W1, W2, Y, colors='black', linestyles='dashed', linewidths=1)
plt.clabel(cp, inline=1, fontsize=10)
cp = plt.contourf(W1, W2, Y, )
plt.xlabel('W1')
plt.ylabel('W2')
plt.show()
```
